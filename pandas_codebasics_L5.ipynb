```python
import pandas as pd
#df = pd.read_csv("weather_data (2).csv")
df = pd.read_csv("weather_data (2).csv", parse_dates=["day"])
#df.set_index('day', inplace=True)
df
```

```python
#we use fillna and dropna to organize those elements which contain null values
```

```python
type(df.day)
```

```python
#type(df.day[0]) gives result as str
#after (parse_dates=["day"]) we get timestamps
type(df.day[0])
```

```python
#df.fillna(0, inplace=True) gives us the change in the original file
#here we choose to
# bring the change that is replace N.A. values in another new copy. This copy we name new_
new_df = df.fillna(0)
new_df
```

fillna(0, inplace=True) or
new_df = new_df.fillna(0) .... these 2 can be used to alter the values to the value of our choice. But we can use
(.....na_values= [])
for substituting whatever values are there with NaN
so these are basically opposites

```python
new_df = df.fillna(
    {
        'temperature' : 0,
        'windspeed' : 0,
        'event' : 'no event'
    }
)
new_df
```

```python
#fills the previous day's values... ie ffill fills forward
new_df = df.fillna(method="ffill")
new_df
```

```python
#fills the next day's value
#bfill fills backward
new_df = df.fillna(method="bfill")
new_df
```

```python
#previously we were copying values from the next row or the previous row... but now, we copy it across columns
new_df = df.fillna(method="bfill", axis="columns")
#limit parameter can be used to control how many times thevalue will be copied backward or forward (...limit=2)
new_df
```

```python
new_df = df.interpolate()
new_df
#df.interpolate(method='')..... this method is by default linear, but it can be quadratic or cubic
#we interpolate a value based on already known data..... method="time"... on the basis of time interpolation occurs
```

```python
#whichever row had any na value gets dropped or discarded if use dropna
new_df = df.dropna()
new_df
```

```python
new_df=df.dropna(how="all")
new_df
#ie in a particular row, all columns have to be na ... only then it is dropped... else it is retained
```

```python
new_df = df.dropna(thresh=1)
#i need 2 vlid values in order to keep a particular row
#if i have more na vales and < 2 valid values, the row gets dropped
new_df
```
