```python
import pandas as pd
import numpy as np
df = pd.read_csv('weather_data (4).csv')
df
```

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

```python
new_df = df.replace([-99999,0],np.NaN)
new_df
```

```python
new_df = df.replace({
   'temperature':-99999,
   'windspeed':-99999,
   'event':'0'
}, np.NaN)
new_df
```

```python
new_df = df.replace(
    {
        -99999: np.NaN,
        'No Event': 'Sunny'
    }
)
new_df
```

```python
#regex... whenever i see a character between A to Z, it is replaced
#regex=True is another condition
new_df = df.replace('[A-Za-z]','',regex=True)
new_df
```

```python
new_df = df.replace(
    {
      'temperature' : 'A-Za-z' ,
      'windspeed' : 'A-Za-z' ,
    },'',regex=True
)
new_df
```

```python
#first list contains what you want to replace
#second list contains the values you want to replace it with
new_df = df.replace(['Rain','Sunny','Snow'],['Foul','Pleasant','Cold'])
new_df
```
